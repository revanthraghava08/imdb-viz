# ***1.Line Plot — trends over time:***
---

```python
import matplotlib.pyplot as plt
fig, ax=plt.subplots()
ax.plot([1, 2, 3, 4], [10, 20, 15, 40], color='cyan', linewidth=2, marker='o')
ax.set_title("Monthly Revenue")
ax.set_xlabel("Month")
ax.set_ylabel("Revenue")
plt.show()
```

# ***3.Histogram — distribution of one variable***
---

```python
import numpy as np
data = np.random.normal(loc=60, scale=15, size=1000) # Mean (average) = 60, Standard deviation = 15, Generate 1000 numbers
fig, ax=plt.subplots()
ax.hist(data, bins=30, color='mediumseagreen', edgecolor='white') # bins -> Divides data into 30 intervals
ax.set_title('Exam Score Distribution')
plt.show()
```

# ***4.Scatter Plot — relationship between two variables***
---

```python
x = np.random.rand(100) * 10
y = x * 2 + np.random.randn(100) * 3
fig, ax=plt.subplots()
ax.scatter(x, y, color='coral', edgecolor='white', alpha=0.6)
ax.set_title('Hours - Studied vs Score')
plt.show()
```

# ***6.Subplots — multiple plots in one figure***
---

```python
fig, axes=plt.subplots(1, 2, figsize=(12,4))
axes[0].plot([1, 2, 3], [4, 5, 6], color='cyan', marker='o')
axes[0].set_title('Plot A')
axes[1].bar(['X', 'Y', 'Z'], [3, 7, 5], color='coral', edgecolor='white')
axes[1].set_title('Plot B')
plt.tight_layout()
plt.show()
```

# ***Exercise 1 — Line Chart: Movie Counts Per Decade***
---

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('imdb_top_1000_cleaned.csv')
decade_counts = df.groupby('Decade')['Series_Title'].count()
fig, ax=plt.subplots(figsize=(10, 5))
ax.plot(decade_counts.index, decade_counts.values, color='cyan', linewidth=2.5, marker='o', markersize=7)
ax.set_title('Number of IMDb Top 1000 Movies Per Decade', fontsize=12)
ax.set_xlabel('Decade')
ax.set_ylabel('Number of Movies')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('movies_per_decade.png', dpi=150, bbox_inches='tight')
plt.show()
```

---
# ***Exercise 2 — Bar Chart: Top 10 Genres by Average Rating***
---

```python
df_genre = df.assign(Genre=df['Genre'].str.split(', ')).explode('Genre')
genre_rating = (df_genre.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).head(10))
fig, ax=plt.subplots(figsize=(10,6))
ax.bar(genre_rating.index, genre_rating.values, color='steelblue', edgecolor='white', width=0.6)
ax.set_title('Top 10 Genres by Average IMDb Rating', fontsize=14)
ax.set_xlabel('Genre')
ax.set_ylabel('Average Rating')
ax.tick_params(axis='x', rotation=45)
ax.set_ylim(7.5, 9.0)
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('top_genres_rating.png', dpi=150, bbox_inches='tight')
plt.show()
```

# ***Exercise 3 — 1×2 Subplot: Ratings Histogram + Rating vs Votes Scatter***
---

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
axes[0].hist(df['IMDB_Rating'], bins=20, color='mediumseagreen', edgecolor='white')
axes[0].set_title('Distribution of IMDb Ratings', fontsize=14)
axes[0].set_xlabel('Rating')
axes[0].set_ylabel('Number of Movies')
axes[0].grid(True, alpha=0.3)
axes[1].scatter(df["No_of_Votes"], df["IMDB_Rating"], alpha=0.4, color='coral', edgecolor='none', s=15)
axes[1].set_title('IMDb Rating vs Number of Votes', fontsize=14)
axes[1].set_xlabel('Number of Votes')
axes[1].set_ylabel('IMDb Rating')
axes[1].grid(True, alpha=0.3)
axes[1].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"{x/1_00_000:.1f}M"))
plt.tight_layout()
plt.savefig('ratings_subplot.png', dpi=150, bbox_inches='tight')
plt.show()
```
